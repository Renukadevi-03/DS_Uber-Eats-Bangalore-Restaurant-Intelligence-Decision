```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/Uber_Eats_data.csv")
```

```python
duplicates = data[data.duplicated(subset=['name', 'location'], keep=False)]
duplicates.sort_values(by=['name', 'location'])
```

```python
data = data.drop_duplicates()
```

```python
data.shape
```

```python
data['rate'] = data['rate'].astype(str).str.extract(r'(\d+\.?\d*)')
data['rate'] = data['rate'].astype(float)
```

```python
data['rate'].dtype
```

```python
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].astype(str).str.replace('₹', '')
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].str.replace(',', '')
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].astype(float)
```

```python
data.rename(columns={
    'rate': 'rating',
    'approx_cost(for two people)': 'cost',
}, inplace=True)
```

```python
data.to_csv("cleaned_uber_eats_data.csv", index=False)
```
